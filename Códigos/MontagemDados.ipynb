1 - contém todos os snps dos indivíduos (snps.csv)  
2 - contém todos os id dos indivíduos (individuos.csv)

```
#1
import pandas as pd
import os
folders = os.listdir('/content/drive/My Drive/ic/Dados - Indivíduos SNPs/adni_gwas_v2')
folders.sort()
```

```
final = []
for folder in folders:
  lista = os.listdir(f'/content/drive/My Drive/ic/Dados - Indivíduos SNPs/adni_gwas_v2/{folder}') #procura mais profundamente todos os .csv em todas as pastas indicadas
  for id in lista:
    final.append((id.split('.'))[0])
print('ids dos individuos: ', final)
print('qtd de ids: ',len(final))
```

```
#to csv
df = pd.DataFrame({'ids':final})
df.to_csv('/content/drive/MyDrive/ic/Dados limpos/ids.csv')
```
