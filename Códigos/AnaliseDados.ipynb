```
import os
folders = os.listdir('/content/drive/My Drive/bkzin/ic/Dados - Indivíduos SNPs/adni_gwas_v2')
folders.sort()
folders
```

```
diferentes = 0 
iguais = 0
for folder in folders:
  file = os.listdir(f'/content/drive/My Drive/ic/Dados - Indivíduos SNPs/adni_gwas_v2/{folder}') #procura mais profundamente todos os .csv em todas as pastas indicadas
  file.sort()
  for id in file: #pego cada um dos cerca de 200 arquivos
    teste = pd.read_csv(f'/content/drive/My Drive/ic/Dados - Indivíduos SNPs/adni_gwas_v2/{folder}/{id}',usecols=["SNP Name"]) 
    indexador = df1["SNP Name"].isin(teste["SNP Name"])
    if (indexador.sum() != teste["SNP Name"].count()): 
      diferentes +=1
    else:
      iguais += 1 
```

```
print(f'Iguais: {iguais} \nDiferentes: {diferentes}')
```

**Portanto todos os individuos têm os mesmos SNPs**

```
df_all_ids = pd.DataFrame()
df_all_ids["SNP Name"] = None
count = 0 
```

```
for folder in folders[4:8]:
  file = os.listdir(f'/content/drive/My Drive/bkzin/ic/Dados - Indivíduos SNPs/adni_gwas_v2/{folder}') #procura mais profundamente todos os .csv em todas as pastas indicadas
  file.sort()
  for id in file: #pego cada um dos cerca de 200 arquivos
    juntar = pd.read_csv(f'/content/drive/My Drive/bkzin/ic/Dados - Indivíduos SNPs/adni_gwas_v2/{folder}/{id}',usecols=["SNP Name","Allele1 - AB","Allele2 - AB"])
    juntar.rename(columns = {"Allele1 - AB":f"Allele1 - AB/{(id.split('.'))[0]}","Allele2 - AB":f"Allele2 - AB/{(id.split('.'))[0]}"},inplace=True) 
    df_all_ids = pd.merge(df_all_ids, juntar, on="SNP Name", how='outer')
    count += 1
    print(f"count {count} \nid {id}")
```
