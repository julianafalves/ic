```
import os
folders = os.listdir('/content/drive/My Drive/ic/Dados - Indivíduos SNPs/adni_gwas_v2')
folders.sort()
folders
```

```
diferentes = 0 
iguais = 0
for folder in folders:
  file = os.listdir(f'/content/drive/My Drive/ic/Dados - Indivíduos SNPs/adni_gwas_v2/{folder}') #procura mais profundamente todos os .csv em todas as pastas indicadas
  file.sort()
  for id in file: #pego cada um dos cerca de 200 arquivos
    teste = pd.read_csv(f'/content/drive/My Drive/ic/Dados - Indivíduos SNPs/adni_gwas_v2/{folder}/{id}',usecols=["SNP Name"]) 
    indexador = df1["SNP Name"].isin(teste["SNP Name"])
    if (indexador.sum() != teste["SNP Name"].count()): 
      diferentes +=1
    else:
      iguais += 1 
```

```
print(f'Iguais: {iguais} \nDiferentes: {diferentes}')
```

**Portanto todos os individuos têm os mesmos SNPs**
