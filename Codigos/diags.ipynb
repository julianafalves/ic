```
import numpy as np
diag_ids = df['Subject'].values.tolist() #identidades relacionados ao seu diagnostico
diag_results = df['Group'].values.tolist()
lprincipal = principal['Unnamed: 0'].tolist()

diags_final = []
for i in lprincipal:
  if i in diag_ids:
    diags_final.append(diag_results[diag_ids.index(i)])
  else:
    diags_final.append(np.nan)
diags_final
```
