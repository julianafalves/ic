```
snpsdict = {}
quantidade_2 = 0
snpslist = []
for gene in ListaGenes:
  handle = Entrez.esearch(db="snp", term=f"{gene.lower()} AND human ", idtype="acc")
  record = Entrez.read(handle)
  count = record['Count']
  handle = Entrez.esearch(db="snp", term=f"{gene.lower()} AND human ", idtype="acc",retmax=count)
  record = Entrez.read(handle)
  handle.close()
  print(gene)
  print(len(record["IdList"]))
  snpsdict[f'{gene}']= record["IdList"]
  quantidade_2 += len(record["IdList"])
  snpslist += record["IdList"]
```

```
df = pd.DataFrame.from_dict(snpsdict, orient='index')
df
```
